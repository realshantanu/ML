**Plotting Scatter plot.**

```python
df1 = data[data["AHD"] == 0]
df2 = data[data["AHD"] == 1]
```

```python
plt.xlabel("Age")
plt.ylabel("RestBP")
plt.title("AHD based on AGE vs RestBP")
plt.scatter(df1["Age"],df1["RestBP"],marker="o",s = 30,color='r',label="NO AHD")
plt.scatter(df2["Age"],df2["RestBP"],marker="*",s=50,color='b',label="AHD")
plt.legend()
plt.show()
```

```python
plt.xlabel("Ca")
plt.ylabel("ExAng")
plt.title("AHD based on Ca vs ExAng")
plt.scatter(df1["Ca"],df1["ExAng"],marker="o",s = 30,color='r',label="NO AHD")
plt.scatter(df2["Ca"],df2["ExAng"],marker="*",s=50,color='b',label="AHD")
plt.legend()
plt.show()
```
